```python
import requests
import pandas as pd
```

```python
url = "https://data.binance.com/api/v3/ticker/24hr"
response = requests.get(url)
```

```python
if response.status_code == 200:
  data = response.json()

  df = pd.DataFrame(data)
    
  df.to_csv("binance_data.csv", index=False)

  print(f"Data successfully saved to binance_data.csv")

else:
  print("Data request is unsuccessful")
```
